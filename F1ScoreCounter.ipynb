````python
def parse_bot(x):
    ans = []
    a = x[3:-3].split('", "')
    for i in a:
        b = i.split('": "')
        #print(b)
        ans.append(b)
    return ans
````

````python
def is_json(myjson):
  try:
    json.loads(myjson)
  except ValueError as e:
    return False
  return True
````

````python
def check_output(x, y):
    x_labels = []
    #print(x)
    data = parse_bot(x)
    for i in data:
        #print(i[0], i[1])
        x_labels.append(i[0])

    #print('===')

    y_labels = []
    same_labels = []
    if is_json(y):
        data2 = json.loads(y)
        for i in data2:
            #print(i)
            y_labels.append(i)
            if i in x_labels:
                same_labels.append(i)
    else:
        #print('YEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEE')
        data2 = parse_bot(y)
        for i in data2:
            #print(i)
            y_labels.append(i)
            if i in x_labels:
                same_labels.append(i[0])
    return same_labels
````

````python
def check_output(x, y):
    x_labels = []
    #print(x)
    data = parse_bot(x)
    for i in data:
        #print(i[0].lower(), i[1])
        x_labels.append(i[0].lower())

    #print('===')

    if y.startswith('```json') and y.endswith('```'):
        y = y[7:-3]
    
    y_labels = []
    same_labels = []
    if is_json(y):
        data2 = json.loads(y)
        for i in data2:
            #print(i)
            y_labels.append(i.lower())
            if i.lower() in x_labels:
                same_labels.append(i.lower())
    else:
        #print('YEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEE')
        data2 = parse_bot(y)
        for i in data2:
            #print(i)
            if type(i)==list:
                continue
            y_labels.append(i.lower())
            if i.lower() in x_labels:
                same_labels.append(i[0].lower())
    return same_labels
````

````python
def check_tag(s):
    tag = ''
    if s[:3] == '<@#':
        tag = s[3:]
        if '>' in tag:
            tag = tag[:tag.find('>')]
        else:
            return 'Not a tag'
    else:
        return 'Does not start with a tag'

    value = ''
    if len(s.split('>')) > 1:
        value = s.split('>')[1]
        if value[-2:] == '<@':
            value = value[:-2]
        else:
            return 'Does not end properly'
    return (tag, value)
````

````python
def standart_size(a):
  package_size_atrs = ['Длина упаковки','Ширина упаковки','Высота упаковки']
  item_size_atrs = ['Длина предмета','Ширина предмета','Высота предмета','Глубина предмета', 'Толщина предмета']
  standart_size = ['(длина)', '(ширина)', '(высота)', '(глубина)', '(толщина)']

  atr_names = [x[0] for x in a]

  package_size = ''
  item_size = ''
  for i in range(len(package_size_atrs)):
    if package_size_atrs[i] in atr_names:
      package_size += a[find_in_arr(a, package_size_atrs[i])][1] + " " + standart_size[i] + ", "
      a.remove(a[find_in_arr(a, package_size_atrs[i])])
  if package_size != '':
    package_size = package_size[:-2]
    a.append(('Габариты упаковки', package_size))

  for i in range(len(item_size_atrs)):
    if item_size_atrs[i] in atr_names:
      item_size += a[find_in_arr(a, item_size_atrs[i])][1] + " " + standart_size[i] + ", "
      a.remove(a[find_in_arr(a, item_size_atrs[i])])
  if item_size != '':
    item_size = item_size[:-2]
    a.append(('Габариты предмета', item_size))

  return a

def find_in_arr(a, x):
  for i in range(len(a)):
    if a[i][0] == x:
      return i
````
