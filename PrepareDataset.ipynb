```python
def check_tag(s):
    tag = ''
    if s[:3] == '<@#':
        tag = s[3:]
        if '>' in tag:
            tag = tag[:tag.find('>')]
        else:
            return 'Not a tag'
    else:
        return 'Does not start with a tag'

    value = ''
    if len(s.split('>')) > 1:
        value = s.split('>')[1]
        if value[-2:] == '<@':
            value = value[:-2]
        else:
            return 'Does not end properly'
    return (tag, value)
```

```python
def inside_tag(s, y, x, start):
    if '@' in x or '@' in y:
        return True
    for i in range(start, len(s)-2):
        if s[i] == '<':
            if s[i+1] == '@' and s[i+2] == '>':
                return True
            break
        if s[i] == '>':
            return True

    for i in range(start, 0, -1):
        if s[i] == '<':
            return True
        if s[i] == '>':
            if s[i-1] != '@':
                return True
            break
    return False
```

```python
import re

def mark_existing_tags(i):
    s = texts[i]

    for tag in tags_d[i]:
        S = s
        
        #print('?',tag)
        values = tag[1]
        if len(values.split(';'))>1:
            values = values.split(';')
        else:
            values = [values]    
        needed_indexes = []
        for v in values: 
            if v == '':
                continue
            if '@' in tag[0]:
                return s
            indexes = list(find_all(s, v))# [m.start() for m in re.finditer(v, s)]
            #print('?!',indexes)
            for index in indexes:
                if not inside_tag(s, tag[0], v, index):
                    needed_indexes.append((v, index))
                #print('?!@',inside_tag(s, tag[0], v, index))
        #for v in needed_indexes:
        needed_indexes = sorted(needed_indexes, key=lambda tup: tup[1])
        #print('!', needed_indexes)
        ans = ''
        prev = 0
        for ind in needed_indexes:
            ans += S[prev:ind[1]] + '<@#' + tag[0] + '>' + ind[0] + '<@>'
            prev = ind[1] + len(ind[0])
        ans += S[prev:]
        s = ans
    return s
```

```python
def find_all(a_str, sub):
    start = 0
    while True:
        start = a_str.find(sub, start)
        if start == -1: return
        yield start
        start += len(sub) # use start += 1 to find overlapping matches

list(find_all('spam spam spam spam', 'spam')) # [0, 5, 10, 15]
```

```python
def get_bio_tags(I):
    a = tokenizer.tokenize(data_text_tags[I])
    b = tokenizer.tokenize(data_text[I])
    
    bio_tags = []
    
    j = 0
    s_name = ''
    f = 'O'
    for i in range(len(a)):
        ###
        #print('\t\t\t',a[i],b[j],i)
        if j>= len(b):
            #print('YOOOOOOOO')
            return bio_tags
        if a[i]==b[j] or '##'+a[i]==b[j]:
            #print(a[i], b[j], f, sep='\t')
            j+=1
            if s_name!='' and s_name!='<@>':
                f = get_s_name(s_name)
                #print(s_name, '='*10)
            if s_name=='<@>':
                f = 'O'
            bio_tags.append(f)
            if f == get_s_name(s_name) and f != 'O':
                f = 'I'+f[1:]
            s_name = ''
            continue
        else:
            #print(a[i])
            s_name+=str(a[i])
            f = 'O'
    return bio_tags
```

```python
def get_s_name(s):
    if s in ['<@#модель>']:
        return 'B-MODEL'
        
    if s in ['<@#странапроизводства>']:
        return 'B-COUNTRY'
        
    if s in ['<@#глубинапредмета>']:
        return 'B-XITEM'
        
    if s in ['<@#высотапредмета>']:
        return 'B-YITEM'
        
    if s in ['<@#ширинапредмета>']:
        return 'B-ZITEM'
        
    if s in ['<@#длинаупаковки>']:
        return 'B-XPACK'
        
    if s in ['<@#высотаупаковки>']:
        return 'B-YPACK'
        
    if s in ['<@#ширинаупаковки>']:
        return 'B-ZPACK'
        
    if s in ['<@#материалкорпуса>']:
        return 'B-MATERIAL'
        
    if s in ['<@#комплекта##ция>']:
        return 'B-COMPLECT'
        
    if s in ['<@#цвет>']:
        return 'B-COLOR'
        
    if s in ['<@#операционнаясистема>']:
        return 'B-OS'
        
    if s in ['<@#процессор>']:
        return 'B-PROCESSOR'

    if s in ['<@#типвидеока##рты>']:
        return 'B-VIDEO'
        
    if s in ['<@#объемпамятивидеока##рты>', '<@#объемоперативно##ипамяти(г##б)>']:
        return 'B-RAM'
        
    if s in ['@#количествоядерпроцессора>']:
        return 'B-KERNELS'
        
    if s in ['<@#объемнакопи##теляs##s##d>']:
        return 'B-SSD'

    if s in ['<@#максимальнаядиагона##льноутбука>', '<@#диагона##ль>']:
        return 'B-DIAGONAL'

    return 'O'
```
